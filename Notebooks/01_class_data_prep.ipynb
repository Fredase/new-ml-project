```python
# Define the path to the data sub-folder
data_folder = project_root / 'data'/'raw'

# Define the path to the CSV file
file_path = data_folder / 'classification_data.csv'

# Read the data using pandas
df = pd.read_csv(file_path)

# Display the data
df.head(10)
```

```python
df_1 = df.copy()

fields = ['label', 'Account type', 'Payment Status of Previous Credit',
         'Purpose', 'Savings type', 'Marital Status', 'Most valuable available asset',
         'Type of apartment', 'Occupation', 'Telephone', 'Foreign Worker']

df_1[fields] = df_1[fields].astype(str)

df_1.info()
```

```python
df_1['Instalment per cent'] = df_1['Instalment per cent']/100

df_1.head(10)
```

```python
## Specify the folder and file path
folder_path = project_root/'data'/'processed'
file_path = os.path.join(folder_path, 'class_data.csv')

# Create the folder if it doesn't exist
os.makedirs(folder_path, exist_ok=True)

# Write the DataFrame to a CSV file in the new folder
df_1.to_csv(file_path, index=False)
```
